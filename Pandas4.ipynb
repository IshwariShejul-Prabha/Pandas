```python
# Pandas Training - Day 4 Assignment 
# Topics: Sorting & Grouping 
 
import pandas as pd

Employee = {
    "Name" : ["Ishwari","Rushikesh","atharv","Dipak","Jyoti"],
    "Department" : ["HR","IT","HR","Finance","IT"],
    "Salary" : [50000,40000,55000,45000,60000]
}

df = pd.DataFrame(Employee)
print(df)

#Sort salary in ascending order. 
df.sort_values("Salary")

#Sort salary in descending order.
df.sort_values("Salary",ascending=False)

#Display unique departments. 
df["Department"].unique()    

#Count employees in each department. 
print(df.groupby("Department")["Name"].count())

#Find average salary of each department. 
print(df.groupby("Department")["Salary"].mean())

#Find maximum salary of each department.
print(df.groupby("Department")["Salary"].max())

# ** QUESTIONS :

# Q1. What is the purpose of sort_values()?
# Ans: The sort_values() function is used to sort the data in a DataFrame by one or more columns in ascending or descending order.

# Q2. What is the difference between ascending and descending sorting?
# Ans: - Ascending (ascending=True): Sorts data from smallest to largest or A to Z.
#      - Descending (ascending=False): Sorts data from largest to smallest or Z to A.
    
# Q3. What is the purpose of unique()?
# Ans: The unique() function returns all the distinct (unique) values from a column without duplicates.

# Ex.,
# City
# Pune
# Mumbai
# Nashik

# Q4. What is the purpose of value_counts()?

# Ans: The value_counts() function counts the frequency of each unique value in a column.

# Ex.,

# City
# Pune      6
# Mumbai    4
# Nashik    3

# Q5. What is groupby()?

# Ans: The groupby() function is used to group data based on one or more columns and perform operations such as sum, mean, count, min, and max on each group.

# Ex., Group students by Course and calculate the average marks.

# Q6. Write the aggregation functions available in Pandas.
# Ans: Common aggregation functions in Pandas are:
#   sum()
#   mean()
#   count()
#   min()
#   max()
#   median()
#   std() (Standard Deviation)
#   var() (Variance)
#   prod() (Product)
#   first()
#   last()
#   size() (Number of rows in each group)
```
