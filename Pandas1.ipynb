```python
# Pandas Training - Day 1 Assignment 
# Topics: Introduction to Pandas, Series, DataFrame, CSV Files

import pandas as pd
student = {
    "Name" : ["Ishwari","Atharv","Rushikesh","Jyoti","Dipak"],
    "Age" : [19,20,22,35,40],
    "City" : ["Pune","Mumbai","Parbhani","Nashik","Pune"],
    "Marks" : [84,65,91,76,83]
}
 
df = pd.DataFrame(student)

# Display DataFrame
print("Student DataFrame")
print(df)

# Save DataFrame into CSV file
df.to_csv("students.csv", index=False)
print("\nCSV file saved successfully!")

# Read CSV file
data = pd.read_csv("students.csv")

# Display loaded data
print("\nLoaded Data")
print(data)

# Questions :

# Q1. What is Pandas? 
# Ans: - Pandas is an open-source Python library used for data manipulation, analysis, and handling structured data. 
#      - It provides two main data structures: Series and DataFrame.

# Q2. Why is Pandas used in AI/ML?
# Ans: Pandas is used in AI/ML because it helps to:
#         - Load datasets
#         - Clean data
#         - Handle missing values
#         - Remove duplicates
#         - Filter and sort data
#         - Prepare data before training machine learning models

# Q3. What is a Series?
# Ans: A Series is a one-dimensional labeled array that can store a single column of data such as integers, strings, or floats.

# ex., 
import pandas as pd

print("\n Series")
data = pd.Series([10, 20, 30, 40])
print(data)

# Q4. What is a DataFrame?
# Ans: - A DataFrame is a two-dimensional table consisting of rows and columns. 
#      - It can store different types of data in different columns.

# Q5. What is a CSV file?
# Ans: - A CSV (Comma-Separated Values) file is a text file used to store tabular data. 
#      - Each row represents a record, and each value is separated by a comma.
# ex.,
# Name,Age,City
# Raj,21,Pune
# Priya,20,Nashik

# Q6. Write a program to create a DataFrame from a dictionary.
# Ans: 
import pandas as pd

student = {
    "Name": ["Raj", "Priya", "Amit"],
    "Age": [21, 20, 22],
    "City": ["Pune", "Nashik", "Mumbai"]
}

df = pd.DataFrame(student)
print(df)

```
