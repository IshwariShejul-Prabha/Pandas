```python
# Pandas Training - Day 3 Assignment 
# Topics: Data Cleaning 

import pandas as pd

data = {
    "Name" : ["Ishwari","Rushii","Atharv","Riya","Rushii"],
    "Age" : [19,20,22,None,14],
    "Marks" : [84,None,78,65,None],
    "Course" : ["AIML","Python","Data Science","IT","Python Full Stack"]
}

df = pd.DataFrame(data)
print(df)

# Find missing values. 
df.isnull()

# Count missing values. 
df.isnull().sum()

# Replace missing values with 0. 
df.fillna(0)

# Replace missing age with average age. 
avg_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(avg_age)
print(df)

# Identify duplicate records. 
df.duplicated()

Remove duplicate records. 
df.drop_duplicates()

# ** QUESTIONS :

# Q1. What is Data Cleaning?
# Ans:   Data Cleaning is the process of identifying and correcting errors in a dataset, 
#        such as missing values, duplicate records, incorrect data types, and inconsistent data, to improve data quality.

# Q2. What is a Missing Value?
# Ans: - A Missing Value is a value that is not available or is empty in a dataset. 
#      - In Pandas, missing values are represented as NaN (Not a Number).

# Q3. What is the purpose of isnull()?
# Ans: - The isnull() function is used to identify missing values in a DataFrame or Series.
#      - It returns True for missing values and False for non-missing values.

# Q4. What is the purpose of fillna()?
# Ans: The fillna() function is used to replace missing values with a specified value, 
#      such as the mean, median, mode, or any custom value.

# Q5. What is the difference between fillna() and dropna()?
# Ans:
# fillna() :	
#     - Replaces missing values with a specified value.	
#     - Keeps all data in the dataset.	
#     - Used when you want to preserve data.	

# dropna() :
#     - Removes rows or columns containing missing values.
#     - Reduces the size of the dataset by deleting data.
#     - Used when missing data is not useful or cannot be filled.

# Q6. What is the purpose of drop_duplicates()?
# Ans: - The drop_duplicates() function is used to remove duplicate rows from a DataFrame, keeping only one copy of each duplicate record. 
#      - It helps improve data quality by eliminating repeated data.

```
