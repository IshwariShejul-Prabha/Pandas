```python
# # PANDAS ASSIGNMNET (06-07-2026)

import pandas as pd

student = {
    "Student_ID": [101,102,103,104,105,106,107,108,109,110],
    "Name": ["Raj","Priya","Amit","Sneha","Rahul","Pooja","Karan","Neha","Raj","Vikas"],
    "Age": [21,20,22,None,21,22,20,21,21,None],
    "Gender": ["Male","Female","Male","Female","Male","Female","Male","Female","Male","Male"],
    "City": ["Pune","Nashik","Mumbai","Pune","Mumbai","Pune","Nashik","Pune","Pune","Mumbai"],
    "Course": ["Python","AI","Python","Data Science","AI","Python","Data Science","AI","Python","Python"],
    "Attendance": [90,95,80,88,None,98,82,91,90,90],
    "Marks": [85,92,75,81,65,96,79,None,85,85],
    "Purchased": ["Yes","Yes","No","Yes","No","Yes","No","Yes","Yes","Yes"]
}

df = pd.DataFrame(student)
print(df)
print("\n")

# Question 1: Replace all missing values in the dataset with 0. 
df.fillna(0)
    
# Question 2: Replace missing values in the Marks column using the average marks. 
avg_marks = df["Marks"].mean()                         
df["Marks"] = df["Marks"].fillna(avg_marks)
print(df["Marks"])
print("\n")
    
# Question 3: Replace missing values in the Age column using the average age. 
avg_age = df["Age"].mean()                                   
df["Age"] = df["Age"].fillna(avg_age)
print(df["Age"])
print("\n")

# Question 4: Remove all rows that contain missing values. 
df.dropna()
print("\n")
    
# Question 5: Find all duplicate records in the dataset. 
print(df.duplicated())
print("\n")
    
# Question 6: Remove duplicate records from the dataset. 
df.drop_duplicates()
print("\n")
    
#Question 7: Display students whose marks are greater than the average marks. 
avg = df["Marks"].mean()
print(df[df["Marks"] > avg])
print("\n")
    
# Question 8: Display students whose attendance is greater than 90%. 
df[df["Attendance"] > 90]
    
# Question 9: Display students whose marks are between 70 and 90. 
df[(df["Marks"] >= 70) & (df["Marks"] <= 90)]
    
# Question 10: Display female students from Pune. 
df[(df["Gender"] == "Female") & (df["City"] == "Pune")]

# Question 11: Display male students whose marks are greater than 80. 
df[(df["Gender"]=="Male") & (df["Marks"]>80)]
    
# Question 12: Find the maximum marks of the students. 
df["Marks"].max()
    
# Question 13: Find the minimum marks of the students. 
df["Marks"].min()
    
# Question 14: Calculate the average marks of all students. 
df["Marks"].mean()
    
# Question 15: Calculate the total marks of all students. 
df["Marks"].sum()
    
# Question 16: Display all unique values from the Course column. 
df["Course"].unique()
    
# Question 17: Count the number of students in each city using value_counts(). 
df["City"].value_counts()
    
# Question 18: Count the number of students in each course. 
df["Course"].value_counts()
    
# Question 19: Group the dataset by City and calculate the average marks. 
df.groupby("City")["Marks"].mean()
    
# Question 20: Group the dataset by Course and calculate: 
# ● Total Marks 
# ● Average Marks 
# ● Minimum Marks 
# ● Maximum Marks 
# ● Total Students 
# using agg(). 
df.groupby("Course").agg({"Marks": ["sum","mean","min","max","count"]})
    
# Question 21: Find the correlation between all numerical columns. 
df.corr(numeric_only = True)
    
# Question 22: Perform Manual Encoding on the Gender column. 
# Example: 
# ● Male → 1 
# ● Female → 0 
df["Gender"] = df["Gender"].replace({"Male":1,"Female":0})

# Question 23: Apply One-Hot Encoding on the following columns: 
# ● City 
# ● Course 
pd.get_dummies(df, columns=["City", "Course"])

# Question 24: Create Features (X) using: 
# ● Age 
# ● Attendance 
# ● Marks 

# Create Target (y) using: 
# ● Purchased 
# Display both X and y.

X = df[["Age", "Attendance", "Marks"]]

y = df["Purchased"]

print("Features (X)")
print(X)

print("\nTarget (y)")
print(y)

df["Purchased"] = df["Purchased"].replace({"Yes":1,"No":0})
```
