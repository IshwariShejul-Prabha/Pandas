```python
# Pandas Training - Day 2 Assignment 
# Topics: Data Selection & Filtering 

import pandas as pd

Employee = {
    "Employee ID" : [101, 102, 103, 104, 105],
    "Name" : ["Ishwari","Rushikesh","Atharv","Dipak","Jyoti"],
    "Department" : ["AIML","Civil","ENTC","Mechanical","Electronics"],
    "Salary" : [25000,30000,100000,4000,40000],
    "City" : ["Pune","Mumbai","Nashik","Pune","Parbhani"]
}

df = pd.DataFrame(Employee)
print(df)

#Display first 3 rows. 
df.head(3)

#Display last 2 rows.
df.tail(2)

#Display dataset shape. 
df.shape

#Display column names.
df.columns

#Display employee records whose salary is greater than 40000. 
df[df["Salary"]>40000]

#Display employees from Pune.
df[df["City"] == "Pune"]

# QUESTIONS :

# Q1. What is the purpose of head()?
# Ans: - The head() function is used to display the first 5 rows of a DataFrame by default. 
#      - It is useful for quickly viewing the beginning of the dataset.

# Q2. What is the purpose of tail()?
# Ans: - The tail() function is used to display the last 5 rows of a DataFrame by default. 
#      - It helps view the end of the dataset.

# Q3. What does shape return?
# Ans: The shape attribute returns the number of rows and columns in a DataFrame as a tuple.

# Ex., (13, 9)

# This means:
# 13 rows
# 9 columns

# Q4. What does info() display?
# Ans: The info() function displays:
#            - Number of rows and columns
#            - Column names
#            - Data types of each column
#            - Non-null (non-missing) values
#            - Memory usage

# Q5. What does describe() display?
# Ans: The describe() function displays summary statistics of numerical columns, 
#      such as: Count, Mean, Median, Standard Deviation (std) ,Minimum (min), Maximum (max)

# Q6. Write a program to display employees whose salary is greater than 50000.
# Ans:
import pandas as pd

employee = {
    "Name": ["Raj", "Priya", "Amit", "Sneha", "Rahul"],
    "Salary": [45000, 60000, 52000, 48000, 70000]
}

df = pd.DataFrame(employee)
print(df[df["Salary"] > 50000])
```
