```python
# Pandas Training - Day 5 Assignment 
# Topics: AI/ML Data Preparation 

import pandas as pd

Customer = {
    "Age" : [19,23,20,35,25,24,19,22,27,28],
    "Salary" : [4000,25000,8000,80000,30000,20000,100000,4000,16000,6000],
    "Gender" : ["Female","Male","Male","Female","Male","Female","Female","Female","Male","Female"],
    "Purchased" : ['Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes', 'No', 'Yes', 'Yes']
}

df = pd.DataFrame(Customer)
print(df)

# Display first 5 rows.
df.head(5)

# Display dataset information.
df.info()

# Display statistical summary. 
df.describe()

# Encode Gender column using map().
df["Gender"] = df["Gender"].map({"Male": 1, "Female": 0})

print("\nAfter Encoding Gender")
print(df)

# Find correlation between Age and Salary
print("\nCorrelation Between Age and Salary")
print(df[["Age", "Salary"]].corr())

# Select Features and Target
X = df[["Age", "Salary", "Gender"]]
y = df["Purchased"]

print("\nFeatures (X)")
print(X)

print("\nTarget (y)")
print(y)

# **QUESTIONS :

# Q1. What is Data Exploration?
# Ans: - Data Exploration is the process of understanding and analyzing a dataset before applying machine learning. 
#      - It includes checking the data, finding missing values, duplicates, data types, and summary statistics.

# Q2. What is a Feature in Machine Learning?
# Ans: - A Feature is an input variable used to train a machine learning model. 
#      - It helps the model make predictions.

# Ex.,
# Age
# Salary
# Gender

# Q3. What is a Target Variable?
# Ans: A Target Variable is the output variable that the machine learning model predicts.

# Ex.,
# Purchased (Yes/No)
# Price
# Marks

# Q4. What is Correlation?
# Ans: - Correlation is a statistical measure that shows the relationship between two numerical variables. 
#      - Its value ranges from -1 to +1.

# +1 → Perfect positive correlation
# 0 → No correlation
# -1 → Perfect negative correlation

# Q5. Why is Encoding required in Machine Learning?
# Ans: - Encoding is required because machine learning algorithms work with numerical data, not text.
#      - Encoding converts categorical values (such as Male/Female or Yes/No) into numerical values.

# Ex.,
# Male → 1
# Female → 0
# Yes → 1
# No → 0

# Q6. What is the difference between Training Data and Testing Data?
# Ans:
# Training Data :
#     - Used to train the machine learning model.	
#     - Used to test the performance of the trained model.
#     - Usually a larger portion of the dataset (e.g., 80%).	
#     - Helps the model learn patterns from the data.	

# Testing Data :
#     - Used to test the performance of the trained model.
#     - Usually a smaller portion of the dataset (e.g., 20%).
#     - Checks how well the model predicts on new, unseen data.
```
