```python
# Pandas-assignment-04-07-2026

# 1. Create a DataFrame containing the following columns: 
# ● Name 
# ● Age 
# ● City 
# ● Marks 
# Add data for 5 students. 
import pandas as pd

student = {
    "Name" : ["Ishwari","Atharv","Rushikesh","Jyoti","Dipak"],
    "Age" : [19,20,22,35,40],
    "City" : ["Pune","Mumbai","Parbhani","Nashik","Pune"],
    "Marks" : [84,65,91,76,83]
}

# 2. Display the complete DataFrame. 
df = pd.DataFrame(student)
print(df)
print("\n")

# 3. Display the first 3 rows using head().
df.head(3)

#4. Display the last 2 rows using tail().
df.tail(2)

# 5. Display the shape of the DataFrame.
df.shape

# 6. Display all column names. 
df.columns

# 7. Display dataset information using info(). 
df.info()
print("\n")

# 8. Display statistical summary using describe().
df.describe()

# 9. Read a CSV file named students.csv.
data = pd.read_csv("students.csv")

# 10. Save the DataFrame into a new CSV file named students_backup.csv. 
df.to_csv("students_backup.csv", index=False)
print("students_backup.csv created successfully!")
print("\n")

# 11. Display only the Name column. 
df["Name"]

# 12. Display Name and Marks columns. 
df[["Name","Marks"]]

# 13. Display students whose marks are greater than 80. 
df[df["Marks"]>80] 

# 14. Display students whose age is greater than 20.
df[df["Age"]>20] 

# 15. Display students who belong to Pune. 
df[df["City"] == "Pune"] 

# 16. Display students whose marks are between 70 and 90.
df[(df["Marks"] >= 70) & (df["Marks"] <= 90)] 

# 17. Display students whose city is Mumbai and marks are greater than 75.
df[(df["City"] == "Mumbai") & (df["Marks"] >= 75)] 

# 18. Sort the DataFrame by Marks in ascending order.
df.sort_values("Marks")      

# 19. Sort the DataFrame by Marks in descending order. 
df.sort_values("Marks",ascending = False)

# 20. Display all unique cities. 
df["City"].unique()

# 21. Count the number of students in each city.
print(df.groupby("City")["Name"].count())

# 22. Display duplicate records.
df.duplicated()  

# 23. Remove duplicate records.
df.drop_duplicates()    

# 24. Check whether the dataset contains missing values. 
df.isnull()
    
# 25. Count missing values in each column.
df.isnull().sum()
```
